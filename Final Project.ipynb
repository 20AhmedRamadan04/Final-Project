Replacing Some Values With Nan 

```python
import numpy as np
Data = Data.replace("Electric Motor" , np.nan)
Data = Data.replace("Electric" , np.nan)
Data = Data.replace("Hybrid" , np.nan)
Data = Data.replace("4.0 (Hybrid)" , np.nan)
Data = Data.replace("2.0 (Electric)" , np.nan)
Data = Data.replace("Hybrid (4.0)" , np.nan)
Data = Data.replace("Electric (100 kWh)" , np.nan)
Data = Data.replace("Electric (93 kWh)" , np.nan)
Data = Data.replace("Electric (tri-motor)" , np.nan)
Data = Data.replace("1.5 + Electric" , np.nan)
Data = Data.replace("Electric (1.5)" , np.nan)
Data = Data.replace("-" , np.nan)
Data = Data.replace("10,000" , np.nan)
Data = Data.replace("1,500" , np.nan)
```

We Will Choose Some Columns To Be Our Dataset Then Train It With A Suitable Algorithm After Splitting It 

So We Will Choose "Engine Size (L)", "Horsepower", "Torque (lb-ft)", And "0-60 MPH Time (seconds)" Columns To Be The Features And "Car Make" Column To Be The Lable

But First We Have To Encode The Data Becuase We Have Object Type Columns So We Should Convert It To Numircal 

```python
from sklearn.preprocessing import LabelEncoder
d_types=Data.dtypes
for i in range(Data.shape[1]):
    if d_types[i]=='object':
        Pr_data = LabelEncoder()
        Data[Data.columns[i]] = Pr_data.fit_transform(Data[Data.columns[i]])
```

After Encoding We can Deal With Our Dataset So First We Will Store Features And Labels Into Variables "X_cat", And "Y_cat" And We Will Split The Data We Choosed With The "train_test_split" Method And Test Size Will Be 0.3 (30%)

But Note That We Will Use The OOP To Be More Organiezed

We Will Make A Class Called Dataset To Be The Parent Class And We Will Define A Function Called Splitting To Split The Data To Train And Test Then We Will Define Another One Called Visualization To Visualize The Data Then We Will Define Another One Called DecisionTree This Is For Modeling And The Last Function Called Accuracy This Is For Measuring The Accuracy Of The Model

```python
class Dataset :
    def __init__(self , X_cat , Y_cat ) :
        self.X_cat = X_cat
        self.Y_cat = Y_cat  
        from sklearn.model_selection import train_test_split
        self.X_cat_train , self.X_cat_test , self.Y_cat_train , self.Y_cat_test = train_test_split(np.array(self.X_cat) , np.array(self.Y_cat) , test_size = 0.3)
        from sklearn.tree import DecisionTreeClassifier 
        self.Model = DecisionTreeClassifier(criterion="entropy")
        self.Model.fit(self.X_cat_train , self.Y_cat_train)
        self.Y_cat_pred = self.Model.predict(self.X_cat_test)
    def Splitting(self) :
            from sklearn.model_selection import train_test_split
            self.X_cat_train , self.X_cat_test , self.Y_cat_train , self.Y_cat_test = train_test_split(np.array(self.X_cat) , np.array(self.Y_cat) , test_size = 0.3)
            print( f"Features Are:\n{self.X_cat}\nLable Is:\n      Car Make\n{self.Y_cat}\nX-cat-train =\n{self.X_cat_train}\nX-cat-test =\n{self.X_cat_test}\nY-cat-train =\n{self.Y_cat_train}\nY-cat-test =\n{self.Y_cat_test}")
    def Visualization(self,corr) :
         import matplotlib.pyplot as plt
         import seaborn as sns
         self.corr = corr
         print(f"Correlation =\n{self.corr}")
         sns.heatmap(self.corr , annot = True)
         plt.show()
         sns.pairplot(self.corr)
         plt.show()
         plt.scatter(self.X_cat.iloc[:,0] , self.Y_cat , color = "blue")
         plt.show()
         plt.bar(self.X_cat.iloc[:,0] , self.Y_cat , color = "blue")
         plt.show()
         sns.catplot(data=Data , kind="count" , x="Engine Size (L)" , hue="Car Make")
         plt.show()
         plt.scatter(self.X_cat.iloc[:,1] , self.Y_cat , color = "red")
         plt.show()
         plt.bar(self.X_cat.iloc[:,1] , self.Y_cat , color = "red")
         plt.show()
         sns.catplot(data=Data , kind="count" , x="Horsepower" , hue="Car Make")
         plt.show()
         plt.scatter(self.X_cat.iloc[:,2] , self.Y_cat , color = "green")
         plt.show()
         plt.bar(self.X_cat.iloc[:,2] , self.Y_cat , color = "green")
         plt.show()
         sns.catplot(data=Data , kind="count" , x="Torque (lb-ft)" , hue="Car Make")
         plt.show()
         plt.scatter(self.X_cat.iloc[:,3] , self.Y_cat , color = "yellow")
         plt.show()
         plt.bar(self.X_cat.iloc[:,3] , self.Y_cat , color = "yellow")
         plt.show()
         sns.catplot(data=Data , kind="count" , x="0-60 MPH Time (seconds)" , hue="Car Make")
         plt.show()
         sns.histplot(self.Y_cat , kde=True)
         plt.show()
    def DecisionTree(self) :
        from sklearn.tree import DecisionTreeClassifier 
        self.Model = DecisionTreeClassifier(criterion="entropy")
        self.Model.fit(self.X_cat_train , self.Y_cat_train)
        self.Y_cat_pred = self.Model.predict(self.X_cat_test)
    def Accuracy(self) :
        from sklearn.metrics import f1_score 
        self.Acc = f1_score(self.Y_cat_test , self.Y_cat_pred , average="micro")
        print(f"The Model Accuracy = {self.Acc}") 
D = Dataset(Data.iloc[:,3:7] , Data.iloc[:,0]) 
D.Splitting()           
D.Visualization(Data.corr())
```
